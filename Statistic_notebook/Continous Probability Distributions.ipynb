```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

# Beta distribution:
## A distribution over a continous variable $\mu \in [0,1]$:
Represent the probability for some binary event. It is governed by two integer parameters $a > 0$ and $b > 0$:<br />
$Beta(\mu\| a, b) = 
\frac{\Gamma(a + b)}{\Gamma(a)\Gamma(b)}
\mu^{a-1}(1-\mu)^{b-1}$

# 
$
Z_n = \frac{X_1 + X_2 + \dots + X_n -n\mu}{\sqrt{n\sigma^2}}=
\sum_{i=1}^{n}\frac{X_i -\mu}{\sqrt{n\sigma^2}}=
\sum_{i=1}^{n}\bar{X_i}
$

```python
M = 10000
a = 1
b = 3
num_sample = [1,2,10,30,100]

for n in num_sample:
    sample_means = []
    for m in range(M):    
        samples = np.random.uniform(a,b,n)
        sample_mean = np.mean(samples)
        sample_means.append(sample_mean)
        
    num_bins = 100
    plt.hist(sample_means, num_bins, alpha=0.5)
    plt.title('N = {}'.format(n))
    plt.show()
```
